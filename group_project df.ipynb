```python
#Visualizations 2018

import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

#Create sample data
Years = [2018]
private_total = [217780]
employment_based = [178350]
direct_purchase = [34846]
tricare = [8537]

# create a bar chart
fig, ax = plt.subplots()
categories = ['Employment Based', 'Direct Purchase', 'Tricare (military)']
values = [employment_based[0], direct_purchase[0], tricare[0]]
ax.bar(categories, values) #breaks here using other method

# add labels and titles
ax.set_xlabel('Categories')
ax.set_ylabel('Number of Insured')
ax.set_title(f'Number of Insured by Category in {Years[0]}')

#change colors
colors = ['blue', 'orange', 'green'] # set colors
plt.bar(categories, values, color=colors) # set axis

## set y-axis tick labels to use commas
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}')) #stack overfolow

plt.show()

# create a pie chart
fig, ax = plt.subplots()
categories = ['Employment Based', 'Direct Purchase', 'Tricare (military)']
values = [employment_based[0], direct_purchase[0], tricare[0]]
ax.pie(values, labels=categories, autopct='%1.1f%%')

# add labels
ax.set_title(f'Number of Insured by Category in {Years[0]}')
ax.legend(labels=categories, loc="top right")
ax.set_xlabel('Categories')
ax.set_ylabel('Number of Insured')

# add title
ax.set_title(f'Percentage of Insured by Category in {Years[0]}')
```
