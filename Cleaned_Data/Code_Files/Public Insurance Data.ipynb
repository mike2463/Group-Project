```python
#2018
medicaretotal_2018 = int(public2018_csv['Medicare Number'].values[0].replace(',', ''))
medicaidtotal_2018 = int(public2018_csv['Medicaid Number'].values[0].replace(',', ''))
VACHAMPVAtotal_2018 = int(public2018_csv['VA/CHAMPVA Number'].values[0].replace(',', ''))
#2019
medicaretotal_2019 = int(public2019_csv['Medicare Number'].values[0].replace(',', ''))
medicaidtotal_2019 = int(public2019_csv['Medicaid Number'].values[0].replace(',', ''))
VACHAMPVAtotal_2019 = int(public2019_csv['VA/CHAMPVA Number'].values[0].replace(',', ''))
#2020
medicaretotal_2020 = int(public2020_csv['Medicare Number'].values[0].replace(',', ''))
medicaidtotal_2020 = int(public2020_csv['Medicaid Number'].values[0].replace(',', ''))
VACHAMPVAtotal_2020 = int(public2020_csv['VA/CHAMPVA Number'].values[0].replace(',', ''))
#2021
medicaretotal_2021 = int(public2021_csv['Medicare Number'].values[0].replace(',', ''))
medicaidtotal_2021 = int(public2021_csv['Medicaid Number'].values[0].replace(',', ''))
VACHAMPVAtotal_2021 = int(public2021_csv['VA/CHAMPVA Number'].values[0].replace(',', ''))
```

```python
insurance_type = ("2018", "2019", "2020", "2021")
insurance_totals = {
    'Medicare': (medicaretotal_2018, medicaretotal_2019, medicaretotal_2020, medicaretotal_2021),
    'Medicaid': (medicaidtotal_2018, medicaidtotal_2019, medicaidtotal_2020, medicaidtotal_2021),
    'VA/CHAMPVA': (VACHAMPVAtotal_2018, VACHAMPVAtotal_2019, VACHAMPVAtotal_2020, VACHAMPVAtotal_2021),
}
```

```python
x = np.arange(len(insurance_type))
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in insurance_totals.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    multiplier += 1


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total')
ax.set_title('Pubic Insurance Type by Year')
ax.set_xticks(x + width, insurance_type)

ax.legend(loc='upper left', ncols=3)

plt.savefig('public_byyear.png')
plt.show()
```
