```python
privatetotal_2018 = int(csv_2018['Private Total Number'].values[0].replace(',', ''))
privatetotal_2019 = int(csv_2019['Private Total Number'].values[0].replace(',', ''))
privatetotal_2020 = int(csv_2020['Private Total Number'].values[0].replace(',', ''))
privatetotal_2021 = int(csv_2021['Private Total Number'].values[0].replace(',', ''))
```

```python
# Plot the charts and apply some styling
plt.plot(years, pubtotals, color="red", label="Public")
plt.plot(years, privtotals, color="green", label="Private")
plt.plot(years, UItotals, color="blue", label="Uninsured")
plt.title("Insurance By Category Over Time")
plt.xlabel("Year")
plt.ylabel("Users")
plt.legend()
plt.savefig ('insurancecompare.png')
plt.show()
```
