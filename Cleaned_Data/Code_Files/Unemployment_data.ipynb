```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
UE = pd.read_csv("Unemployment_rate.csv")
UE
```

```python
UE["Value"].describe()
```

```python
x_axis = UE['Label']

y_axis = UE['Value']
plt.figure(figsize=(20,4))
plt.plot(x_axis,y_axis)
plt.xticks(x_axis, rotation = 90)
plt.ylabel('Unemployment Rate (Percentage %)', loc='center')
plt.xlabel('Month/Year')
```
